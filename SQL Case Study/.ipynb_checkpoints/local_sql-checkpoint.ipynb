```python
import sqlite3
from sqlite3 import Error
import pandas as pd
 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT *
        FROM FACILITIES
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
#connect to the database using the provided methdos
db =  database = "sqlite_db_pythonsqlite.db"
conn = create_connection(database)

#loading the three sheets into pandas DFs 
facilities = pd.read_sql_query('SELECT * FROM Facilities', conn)
members = pd.read_sql_query('SELECT * FROM Members', conn)
bookings = pd.read_sql_query('SELECT * FROM Bookings', conn)
```

***Q10: Produce a list of facilities with a total revenue less than 1000. The output of facility name and total revenue, sorted by revenue. Remember that there's a different cost for guests and members! ***

```python
#initializing counter
counter = {}

#looping over rows in bookings, looking for guests, members, and associating relevant price
for i, b in bookings.iterrows():
    if (b.facid)in counter.keys():
        if b.memid ==0:
            counter[b.facid] += facilities.iloc[b.facid]['guestcost']
        else: 
            counter[b.facid] += facilities.iloc[b.facid]['membercost']
    if (b.facid) not in counter.keys():
        if b.memid ==0:
            counter[b.facid] = facilities.iloc[b.facid]['guestcost']
        else: 
            counter[b.facid] = facilities.iloc[b.facid]['membercost']
 
revenue = {facilities.iloc[k]['name']: v for k, v in counter.items() if v <1000}
rev_pd = pd.Series(revenue).sort_values(ascending=False)
print(rev_pd)
```

***Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order***

```python
mem_rec = members[['surname','firstname','recommendedby']]
recs = mem_rec.replace('',0)

def id_to_name(id):
    """takes member id in, returns the name"""
    id = int(id)
    if id == 0 : return ''
    else:
        mem = members.iloc[id]
        return mem['surname'] + ' ' + mem['firstname']


recs['rec_name'] = recs.apply(lambda row: id_to_name(row['recommendedby']), axis =1)

rec_ord = recs[['surname', 'firstname','rec_name']]
rec_ord = rec_ord[rec_ord['surname'] != 'GUEST']
rec_ord.sort_values(['surname','firstname'], inplace=True)
print(rec_ord)
```

***Q12: Find the facilities with their usage by member, but not guests***


```python
usage = {row[0] : [] for index, row in facilities.iterrows()}


for index, row in bookings.iterrows():
    if row[2] != 0:
        usage[row[1]].append(row[2])

ids = list(set(members.memid))[1:]

use_df = pd.DataFrame()

for i in ids:
    use_df[i] = pd.Series([v.count(i) for k,v in usage.items()])
    
use_df.set_index(facilities.name, inplace=True)
use_df.columns = members.iloc[1:,2] + ' ' + members.iloc[1:,1]
use_df
```

*** Q13: Find the facilities usage by month, but not guests ***

```python
usage_m = {row[0] : [] for index, row in facilities.iterrows()}
usage_m

for index, row in bookings.iterrows():
    if row[2] != 0:
        usage_m[row[1]].append(int(row[3][5:7]))
        
        
use_m_df = pd.DataFrame()

for i in range(1,13):
    use_m_df[i] = pd.Series([v.count(i) for k,v in usage.items()])

use_m_df.set_index(facilities.name, inplace=True)
use_m_df.columns = ['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
use_m_df
```
